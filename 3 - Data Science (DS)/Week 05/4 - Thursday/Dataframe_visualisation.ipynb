```python
#pip install numpy pandas matplotlib seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Load the Dataset

```python
#Check available dataset
#sns.get_dataset_names()

#Built in titanic dataset
df = sns.load_dataset("titanic")

#Fetch data from Github
# url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
# df = pd.read_csv(url)

#Read csv file from local dir
# filename = 'titanic.csv'
# df = pd.read_csv(filename)
```

```python
#Drop null values
new_df = df.dropna()
new_df.info()
```

```python
#Replace null value with mean, median or mode

#Older method
x = df['age'].mean()
#df['age'].fillna(x, inplace = True) #inplace = True overwrites the dataframe

#Preferred method
df['age'] = df['age'].fillna(x)
```

```python
df.info()
```

Visualisation

```python
sns.histplot(df['age'], bins=30, color='g', kde=True) #Kernel density plots
plt.savefig('hist.png')
```

```python
#Removing rows
for x in df.index:
    if df.loc[x,"age"] > 50:
        df.drop(x,inplace=True)
sns.histplot(df['age'], bins=30, color='purple', kde=True)        
```
